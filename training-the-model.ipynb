```python
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F

#(train_X, train_y), (test_X, test_y) = mnist.load_data()
mndata = MNIST('/Users/amre/Documents/GitHub/MNIST/data/')

train_X, labels_train = map(np.array, mndata.load_training())
test_X, labels_test = map(np.array, mndata.load_testing())
train_X = train_X/255.0
test_X = test_X/255.0

X_train_tensor = torch.from_numpy(train_X).float()
y_train_tensor = torch.from_numpy(labels_train).long()

X_test_tensor = torch.from_numpy(test_X).float()
y_test_tensor = torch.from_numpy(labels_test).long()
```

```python
# part a


import torch

# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
H = 128
n, d = X_train_tensor.shape


    

model = torch.nn.Sequential(
    torch.nn.Linear(d, H),
    torch.nn.Linear(H, d),
)

loss_fn = torch.nn.MSELoss(reduction='sum')

learning_rate = 1e-2
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for t in range(5000):
    y_pred = model(X_train_tensor)
    loss = loss_fn(y_pred, X_train_tensor)
    if t % 100 == 99:
        print(t, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
